### PHATE Analysis

Here we repeat the PHATE analysis previously performed on the cells, but also in 3D to better model trajectories in slingshot. 


```python
import matplotlib.pyplot as plt
import pandas as pd
import phate
import seaborn as sns


from scipy.stats import zscore
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

Reading in the data and the optimal features selected in the Stallaert et al., 2022.:

```python
cells = pd.read_csv("/media/omar/1354B7D111621BC0/projects/unc/wayne_rpe/data/full_df.csv")
features = pd.read_csv("/media/omar/1354B7D111621BC0/projects/unc/wayne_rpe/data/final_features.csv")

cells = cells[list(features['Row.names'])]
```

```python
cells.head()
```

Check if values are already Z-score normed (if they are then the mean should be 0 or somewhat close):

```python
cells.mean(axis=0)
```

These values seem to be already Z-score normalized so we can go ahead with PHATE. 

PHATE parameters previously used (from Cell Systems paper):
- knn = 200
- t = 12
- gamma = 1

```python
phate_op = phate.PHATE(n_jobs=2, knn=200, t=12, gamma=1)
cells_phate = phate_op.fit_transform(cells)
```
